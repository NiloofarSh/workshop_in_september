## Getting information for multiple proteins (high-throughput)

Link to [PDBe-KB API](https://www.ebi.ac.uk/pdbe/graph-api/pdbe_doc/)


````python
import requests
import json
import sys

uniprotIds = ['P04637','P02340','P10361','P56423','O09185','Q29537','Q9TUB2','Q9TTA1','P61260','P56424','Q8SPZ3','P79892','P67939','P13481','P67938','Q9WUR6','Q95330','Q00366','P51664','O36006']

# URL to get general protein info from Uniprot
uniprotUrlName = "https://www.ebi.ac.uk/proteins/api/proteins/{}"

# URL to get binding site information from the PDB based on the protein Uniprot ID
pdbeKbUrlName = "https://www.ebi.ac.uk/pdbe/graph-api/uniprot/ligand_sites/{}"


uniprotHeaders = {
    'Content-Type': 'application/json',
    'Cache-Control': 'no-cache'
}

#
# Look for all ligands that bind P53
#

allLigandNames = set()

for uniprotId in uniprotIds:
    print(uniprotId)
    
    # This is the raw text string from the server
    result = requests.get(uniprotUrlName.format(uniprotId),headers=uniprotHeaders)
    
    # We now convert this into Python variables
    currentData = json.loads(result.text)
    
    print(currentData['accession'])
    print(currentData['id'])
    print(currentData['proteinExistence'])
    print(currentData['sequence'])
    
    sequence = currentData['sequence']['sequence']
    print(sequence)

    result = requests.get(pdbeKbUrlName.format(uniprotId),headers=uniprotHeaders)
    
    print(result.text)
    ligandInfo = json.loads(result.text)
    
    # Check all ligands
    if uniprotId in ligandInfo.keys():
        ligandBindingData = ligandInfo[uniprotId]['data']

        for currentLigand in ligandBindingData:
            ligandName = print(currentLigand['name'])

            allLigandNames.add(ligandName)

            # {"name":"6,7-DIHYDRO[1,4]DIOXINO[2,3-F][1,3]BENZOTHIAZOL-2-AMINE","accession":"X0U","residues":[{"startIndex":145,"endIndex":145,"startCode":"LEU","endCode":"LEU","indexType":"UNIPROT","pdbEntries":[{"pdbId":"2x0u","entityId":1,"chainIds":["B","A"]}]}
   
    print(allLigandNames)
````
